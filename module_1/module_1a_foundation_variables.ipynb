# Part 2: The Python Interpreter - Your Personal Translator

## How Code Becomes Execution

Think of the Python interpreter as a **very literal translator**:

```
You write code → Python reads it → Computer executes it
   (English)      (Translation)      (Action)
```

### Key Characteristics:
1. **Line by line**: Python reads from top to bottom
2. **Literal**: It does exactly what you say (no more, no less)
3. **Immediate**: Interpreted languages run instantly (vs compiled languages)

### Compiled vs Interpreted
- **Compiled** (C, Java): Translate entire book first → then read
- **Interpreted** (Python): Translate sentence by sentence as you read

**Advantage of Interpreted**: Instant feedback! You can test code immediately.

# Part 4: Variables - Labels That Point to Values

## The Mental Model

### ❌ Wrong Mental Model: Variables are boxes
Many beginners think: "A variable is a box that stores a value."

### ✅ Correct Mental Model: Variables are labels
A variable is a **name tag** that points to a value in memory.

```
Memory:
  [42] ← label: age
```

This distinction becomes crucial later when we work with lists and objects.

---

## Creating Your First Variable

````python
# Create a variable
age = 25
print(age)
````

### What Happened?
1. `age` is the **variable name** (the label)
2. `=` is the **assignment operator** ("point this label at...")
3. `25` is the **value** (what the label points to)
4. Python created a spot in memory, stored `25`, and labeled it `age`

### Visualize It
👉 [Python Tutor: Variable Assignment](http://pythontutor.com/visualize.html#code=age%20%3D%2025%0Aprint%28age%29&cumulative=false&heapPrimitives=nevernest&mode=edit&origin=opt-frontend.js&py=3&rawInputLstJSON=%5B%5D&textReferences=false)

## Variable Assignment Rules

### Valid Variable Names:
- Must start with a letter or underscore: `age`, `_count` ✅
- Can contain letters, numbers, underscores: `user_name_2` ✅
- **Case-sensitive**: `age` ≠ `Age` ≠ `AGE`

### Invalid Variable Names:
- Can't start with a number: `2fast` ❌
- Can't use special characters: `user-name`, `my.var` ❌
- Can't use Python keywords: `print`, `if`, `for` ❌

### Python Naming Convention (Snake Case):
Use lowercase with underscores for spaces:
- `user_name` ✅ (Python standard)
- `userName` ❌ (JavaScript/Java style)
- `UserName` ❌ (Reserved for classes)

**Why does this matter?** Reading code written by other Python developers (including yourself in 6 months)!

## Multiple Variables

````python
# Create multiple variables
name = "Alice"
age = 25
city = "Stockholm"

print(name)
print(age)
print(city)
````

### Variables Can Change (They're "Variable" After All!)

````python
age = 25
print("Age is:", age)

# One year passes...
age = 26
print("Age is now:", age)
````

**What happened?** The label `age` now points to a different value in memory. The old value (`25`) gets forgotten.

# Part 5: Data Types - The Four Core Types

Python needs to know what **kind** of data you're working with. This affects what you can do with it.

## The Detective Game 🔍

Look at these values. Can you guess their types?

```python
"Hello"      # Type: ?
42           # Type: ?
3.14         # Type: ?
True         # Type: ?
"42"         # Type: ? (Tricky!)
```

Let's check our guesses:

````python
# The type() function reveals the type
print(type("Hello"))
print(type(42))
print(type(3.14))
print(type(True))
print(type("42"))  # Notice the quotes!
````

## The Four Core Types

### 1. `str` (String) - Text Data
- Anything wrapped in quotes
- Single quotes `'Hello'` or double quotes `"Hello"` (your choice!)
- Examples: `"Alice"`, `'Stockholm'`, `"42"`, `"True"`

### 2. `int` (Integer) - Whole Numbers
- Numbers without decimals
- Can be positive, negative, or zero
- Examples: `42`, `-10`, `0`, `1000`

### 3. `float` (Floating-Point) - Decimal Numbers
- Numbers with decimals
- Examples: `3.14`, `-0.5`, `2.0`
- **Note**: Even `2.0` is a float (has a decimal point)

### 4. `bool` (Boolean) - True or False
- Only two values: `True` or `False`
- **Case-sensitive**: `True` ✅, `true` ❌
- Used for decisions (more on this later!)

## Why Do Types Matter?

Different types behave differently:

````python
# Numbers can be added
print(10 + 5)  # What will this print?
````

````python
# Strings can be "added" too (concatenation)
print("Hello" + " " + "World")  # What will this print?
````

````python
# But you can't mix types easily
print("I am " + 25 + " years old")  # What will happen?
````

### 🔥 Gotcha #3: Type Confusion

You got a `TypeError`! Python can't add a string and a number.

**The Fix**: Convert the number to a string first:

````python
# Correct way
print("I am " + str(25) + " years old")

# Or use comma separation (easier!)
print("I am", 25, "years old")
````

## Type Checking: The `type()` Function

When in doubt, check!

````python
age = 25
name = "Alice"
height = 1.65
is_student = True

print("age is type:", type(age))
print("name is type:", type(name))
print("height is type:", type(height))
print("is_student is type:", type(is_student))
````
