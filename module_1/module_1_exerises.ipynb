## Bonus: Gotcha #12 - Floating Point Precision

### Predict: What will this print?


````python
result = 0.1 + 0.2
print(result)
print(result == 0.3)
````

**Surprise!** `0.1 + 0.2 = 0.30000000000000004` - not exactly `0.3`.

Computers store decimals in binary, and some values can't be represented exactly (just like 1/3 = 0.333... in decimal). For most programs this doesn't matter, but for money calculations use Python's `Decimal` module.
