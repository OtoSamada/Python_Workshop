# Variables - Labels That Point to Values

## The Mental Model

### ❌ Wrong Mental Model: Variables are boxes
Many beginners think: "A variable is a box that stores a value."

### ✅ Correct Mental Model: Variables are labels
A variable is a **name tag** that points to a value in memory.

```
Memory:
  [42] ← label: age
```

This distinction becomes crucial later when we work with lists and objects.

![Screenshot 2026-02-16 at 8.08.02 PM.png](<attachment:Screenshot 2026-02-16 at 8.08.02 PM.png>)

---

## Creating Your First Variable

````python
# Create a variable
age = 25
print(age)
````

### What Happened?
1. `age` is the **variable name** (the label)
2. `=` is the **assignment operator** ("point this label at...")
3. `25` is the **value** (what the label points to)
4. Python created a spot in memory, stored `25`, and labeled it `age`

## Variable Assignment Rules

### Valid Variable Names:
- Must start with a letter or underscore: `age`, `_count` ✅
- Can contain letters, numbers, underscores: `user_name_2` ✅
- **Case-sensitive**: `age` ≠ `Age` ≠ `AGE`

### Invalid Variable Names:
- Can't start with a number: `2fast` ❌
- Can't use special characters: `user-name`, `my.var` ❌
- Can't use Python keywords: `print`, `if`, `for` ❌

### Python Naming Convention (Snake Case):
Use lowercase with underscores for spaces:
- `user_name` ✅ (Python standard)
- `userName` ❌ (JavaScript/Java style)
- `UserName` ❌ (Reserved for classes)

**Why does this matter?** Reading code written by other Python developers (including yourself in 6 months)!

## Multiple Variables

````python
# Create multiple variables
name = "Alice"
age = 25
city = "Stockholm"

print(name)
print(age)
print(city)
````

### Variables Can Change (They're "Variable" After All!)

````python
age = 25
print("Age is:", age)

# One year passes...
age = 26
print("Age is now:", age)
````

**What happened?** The label `age` now points to a different value in memory. The old value (`25`) gets forgotten.

# Data Type

Python needs to know what **kind** of data you're working with. This affects what you can do with it.


Look at these values. Can you guess their types?

```python
"Hello"      # Type: ?
42           # Type: ?
3.14         # Type: ?
True         # Type: ?
"42"         # Type: ?
```

Let's check our guesses:

````python
# The type() function reveals the type
print(type("Hello"))
print(type(42))
print(type(3.14))
print(type(True))
print(type("42"))  # Notice the quotes!
````

## The Four Core Types

### 1. `str` (String) - Text Data
- Anything wrapped in quotes
- Single quotes `'Hello'` or double quotes `"Hello"` (your choice!)
- Examples: `"Alice"`, `'Stockholm'`, `"42"`, `"True"`

### 2. `int` (Integer) - Whole Numbers
- Numbers without decimals
- Can be positive, negative, or zero
- Examples: `42`, `-10`, `0`, `1000`

### 3. `float` (Floating-Point) - Decimal Numbers
- Numbers with decimals
- Examples: `3.14`, `-0.5`, `2.0`
- **Note**: Even `2.0` is a float (has a decimal point)

### 4. `bool` (Boolean) - True or False
- Only two values: `True` or `False`
- **Case-sensitive**: `True` ✅, `true` ❌
- Used for decisions (more on this later!)

## Why Do Types Matter?

Different types behave differently:

````python
# Numbers can be added
print(10 + 5)  # What will this print?
````

````python
# Strings can be "added" too (concatenation)
print("Hello" + " " + "World")  # What will this print?
````

````python
# But you can't mix types easily
print("I am " + 25 + " years old")  # What will happen?
````

### Gotcha #3: Type Confusion

You got a `TypeError`! Python can't add a string and a number.

**The Fix**: Convert the number to a string first:

````python
# Correct way
print("I am " + str(25) + " years old")

# Or use comma separation (easier!)
print("I am", 25, "years old")
````

## Type Checking: The `type()` Function

When in doubt, check!

````python
age = 25
name = "Alice"
height = 1.65
is_student = True

print("age is type:", type(age))
print("name is type:", type(name))
print("height is type:", type(height))
print("is_student is type:", type(is_student))
````

# Arithmetic Operations

## Python as Your Super Calculator

Python can do all basic math operations (and much more!).

### The Basic Operators

````python
# Addition
print("10 + 5 =", 10 + 5)

# Subtraction
print("10 - 5 =", 10 - 5)

# Multiplication
print("10 * 5 =", 10 * 5)

# Division
print("10 / 5 =", 10 / 5)
````

## More Powerful Operators

````python
# Exponentiation (power)
print("2 ** 3 =", 2 ** 3)  # 2 to the power of 3
print("10 ** 2 =", 10 ** 2)  # 10 squared

# Integer Division (floor division)
print("17 // 5 =", 17 // 5)  # Gives whole number only

# Modulo (remainder)
print("17 % 5 =", 17 % 5)  # What's left over?
````

### When is Modulo Useful?
- **Check if a number is even**: `number % 2 == 0`
- **Get last digit**: `number % 10`
- **Wrap around**: Think of a clock (13 % 12 = 1)

## Gotcha #6: Division Always Returns a Float

### Predict First: What type will the result be?

````python
result = 10 / 2
print("Result:", result)
print("Type:", type(result))
````

````python
result = 10 // 2
print("Result:", result)
print("Type:", type(result))
````

Remember PEMDAS from school?
- **P**arentheses
- **E**xponents
- **M**ultiplication / **D**ivision (left to right)
- **A**ddition / **S**ubtraction (left to right)

### Predict: What will these print?

### Gotcha #7: Forgetting Operator Precedence

```python

```

**Pro Tip**: When in doubt, use parentheses! They make your intention clear.

# String Operations

## Strings Aren't Just Text - They're Powerful!

### Concatenation: Joining Strings Together

````python
# The + operator joins strings
first_name = "Alice"
last_name = "Johnson"
full_name = first_name + " " + last_name
print(full_name)
````

````python
# Building sentences
greeting = "Hello, " + "my name is " + full_name + "!"
print(greeting)
````

### Repetition: Multiplying Strings

````python
# The * operator repeats strings
print("Ha" * 3)
print("=" * 40)
print("Python! " * 5)
````

**Useful for**:
- Creating separators: `"-" * 50`
- Generating patterns: `"*" * 10`
- Quick formatting: `"=" * 20`

````python
## Gotcha #8: Can't Add Strings and Numbers

### Predict: Will this work?
````

````python
age = 25
message = "I am " + age + " years old"
print(message)
````

**Error!** `TypeError: can only concatenate str (not "int") to str`

### Why?
Python doesn't know if you want:
- `"I am " + "25"` (string concatenation) → `"I am 25"`
- `"I am " + 25` (maybe you meant to add the number?)

So it refuses to guess!

### Three Solutions:

````python
age = 25

# Solution 1: Convert to string with str()
message = "I am " + str(age) + " years old"
print(message)

# Solution 2: Use comma separation (easier!)
print("I am", age, "years old")

# Solution 3: f-strings (we'll learn this later)
print(f"I am {age} years old")
````

# Comparison Operators

## Making Decisions: True or False?

Comparison operators let us ask questions about our data. They always return `True` or `False` (booleans).

### The Six Comparison Operators

````python
x = 10
y = 5

print("x == y:", x == y)  # Equal to
print("x != y:", x != y)  # Not equal to
print("x > y:", x > y)    # Greater than
print("x < y:", x < y)    # Less than
print("x >= y:", x >= y)  # Greater than or equal to
print("x <= y:", x <= y)  # Less than or equal to
````

## Gotcha #9: = vs ==

**This is one of the most common beginner mistakes!**

### `=` is Assignment ("becomes")

````python
x = 10  # x becomes 10
print(x)
````

### `==` is Comparison ("is equal to?")

````python
x == 10  # Is x equal to 10? Returns True or False
print(x == 10)
print(x == 5)
````

**Remember**:
- `=` → Makes a variable
- `==` → Asks a question

## Comparisons with Strings

````python
name1 = "Alice"
name2 = "alice"
name3 = "Alice"

print(name1 == name3)  # True
print(name1 == name2)  # False - case sensitive!
````

**Remember**: String comparisons are **case-sensitive**!
- `"Alice"` ≠ `"alice"`
- `"Python"` ≠ `"python"`

# Part 5: Type Conversion

## Changing Data Types

Python provides functions to convert between types:

### The Conversion Functions

| Function | Purpose | Example |
|----------|---------|----------|
| `int()` | Convert to integer | `int("42")` → `42` |
| `float()` | Convert to float | `float("3.14")` → `3.14` |
| `str()` | Convert to string | `str(42)` → `"42"` |
| `bool()` | Convert to boolean | `bool(1)` → `True` |

## Converting Strings to Numbers

````python
# String to integer
age_text = "25"
age_number = int(age_text)
print("Type before:", type(age_text))
print("Type after:", type(age_number))
print("Can now do math:", age_number + 5)
````

````python
# String to float
price_text = "19.99"
price_number = float(price_text)
print("Type:", type(price_number))
print("Total for 3:", price_number * 3)
````

## Converting Numbers to Strings

````python
# Number to string (for concatenation)
score = 100
message = "Your score is " + str(score) + " points!"
print(message)
````

## Gotcha #11: Conversion Can Fail

### What happens if we try to convert invalid data?

````python
# This will cause an error
int("Hello")
````

````python
# This will also cause an error
int("3.14")
````

**ValueError**: `invalid literal for int() with base 10`

### What's Valid?
- ✅ `int("42")` - valid integer string
- ✅ `float("3.14")` - valid float string
- ✅ `float("42")` - can convert int string to float
- ❌ `int("3.14")` - can't directly convert float string to int
- ❌ `int("Hello")` - not a number at all

**Solution for "3.14" → int**: Convert to float first!
```python
int(float("3.14"))  # Returns 3
```

## Fixing Our Age Program

````python
# Get age from user and convert immediately
age = int(input("How old are you? "))
next_year = age + 1
print("Next year you'll be", next_year)
````

**Pattern**: Convert input immediately if you need a number:
```python
age = int(input("Age: "))           # For whole numbers
height = float(input("Height: "))   # For decimals
name = input("Name: ")              # Keep as string
```

# 💡 Debugging Tips

When your code doesn't work, follow this checklist:

## 1. Read the Error Message
- Start at the **bottom** (the actual error)
- Look at the **line number**
- Read what Python **expected**

## 2. Common Fixes
**TypeError: can only concatenate str**
→ Did you forget `str()` or use `,` instead of `+`?

**ValueError: invalid literal for int()**
→ Is the input actually a number? Try `type()` to check.

**NameError: name 'X' is not defined**
→ Did you spell the variable correctly? Is it defined above this line?

**SyntaxError: invalid syntax**
→ Missing quotes, parentheses, or colon? Check the line above too!

## 3. The Print Debug Method
When stuck, add print statements to see what's happening:

````python
age = input("Age: ")
print("DEBUG - age value:", age)
print("DEBUG - age type:", type(age))
age = int(age)
print("DEBUG - age after conversion:", age, type(age))
````

## Bonus: Gotcha #12 — Floating Point Precision

### Predict: What will this print?


````python
result = 0.1 + 0.2
print(result)
print(result == 0.3)
````

**Surprise!** `0.1 + 0.2 = 0.30000000000000004` — not exactly `0.3`.

Computers store decimals in binary, and some values can't be represented exactly (just like ⅓ = 0.333... in decimal). For most programs this doesn't matter, but for money calculations use Python's `Decimal` module.
