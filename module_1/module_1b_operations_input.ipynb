# Part 1: Arithmetic Operations

## Python as Your Super Calculator

Python can do all basic math operations (and much more!).

### The Basic Operators

````python
# Addition
print("10 + 5 =", 10 + 5)

# Subtraction
print("10 - 5 =", 10 - 5)

# Multiplication
print("10 * 5 =", 10 * 5)

# Division
print("10 / 5 =", 10 / 5)
````

## More Powerful Operators

````python
# Exponentiation (power)
print("2 ** 3 =", 2 ** 3)  # 2 to the power of 3
print("10 ** 2 =", 10 ** 2)  # 10 squared

# Integer Division (floor division)
print("17 // 5 =", 17 // 5)  # Gives whole number only

# Modulo (remainder)
print("17 % 5 =", 17 % 5)  # What's left over?
````

### 💡 When is Modulo Useful?
- **Check if a number is even**: `number % 2 == 0`
- **Get last digit**: `number % 10`
- **Wrap around**: Think of a clock (13 % 12 = 1)

## 🔥 Gotcha #6: Division Always Returns a Float

### 🤔 Predict First: What type will the result be?

````python
result = 10 / 2
print("Result:", result)
print("Type:", type(result))
````

**Surprise!** Even though `10 / 2 = 5`, the result is `5.0` (a float).

**Why?** Division (`/`) always returns a float in Python 3.

**Want an integer?** Use integer division (`//`):

````python
result = 10 // 2
print("Result:", result)
print("Type:", type(result))
````

# Part 2: String Operations

## Strings Aren't Just Text - They're Powerful!

### Concatenation: Joining Strings Together

````python
# The + operator joins strings
first_name = "Alice"
last_name = "Johnson"
full_name = first_name + " " + last_name
print(full_name)
````

````python
# Building sentences
greeting = "Hello, " + "my name is " + full_name + "!"
print(greeting)
````

### Repetition: Multiplying Strings

````python
# The * operator repeats strings
print("Ha" * 3)
print("=" * 40)
print("Python! " * 5)
````

**Useful for**:
- Creating separators: `"-" * 50`
- Generating patterns: `"*" * 10`
- Quick formatting: `"=" * 20`

## 🔥 Gotcha #8: Can't Add Strings and Numbers

### 🤔 Predict: Will this work?

````python
age = 25
message = "I am " + age + " years old"
print(message)
````

**Error!** `TypeError: can only concatenate str (not "int") to str`

### Why?
Python doesn't know if you want:
- `"I am " + "25"` (string concatenation) → `"I am 25"`
- `"I am " + 25` (maybe you meant to add the number?)

So it refuses to guess!

### 💡 Three Solutions:

````python
age = 25

# Solution 1: Convert to string with str()
message = "I am " + str(age) + " years old"
print(message)

# Solution 2: Use comma separation (easier!)
print("I am", age, "years old")

# Solution 3: f-strings (we'll learn this later)
print(f"I am {age} years old")
````

# Part 3: Comparison Operators

## Making Decisions: True or False?

Comparison operators let us ask questions about our data. They always return `True` or `False` (booleans).

### The Six Comparison Operators

````python
x = 10
y = 5

print("x == y:", x == y)  # Equal to
print("x != y:", x != y)  # Not equal to
print("x > y:", x > y)    # Greater than
print("x < y:", x < y)    # Less than
print("x >= y:", x >= y)  # Greater than or equal to
print("x <= y:", x <= y)  # Less than or equal to
````

## 🔥 Gotcha #9: = vs ==

**This is one of the most common beginner mistakes!**

### `=` is Assignment ("becomes")

````python
x = 10  # x becomes 10
print(x)
````

### `==` is Comparison ("is equal to?")

````python
x == 10  # Is x equal to 10? Returns True or False
print(x == 10)
print(x == 5)
````

**Remember**:
- `=` → Makes a variable
- `==` → Asks a question

## Comparisons with Strings

````python
name1 = "Alice"
name2 = "alice"
name3 = "Alice"

print(name1 == name3)  # True
print(name1 == name2)  # False - case sensitive!
````

**Remember**: String comparisons are **case-sensitive**!
- `"Alice"` ≠ `"alice"`
- `"Python"` ≠ `"python"`

# Part 4: Getting User Input

## Making Programs Interactive

So far, all our variables have been hardcoded. Let's make our programs talk to users!

### The `input()` Function

````python
# Get user's name
name = input("What is your name? ")
print("Hello,", name, "!")
````

### How `input()` Works:
1. Displays the prompt (the text in quotes)
2. Waits for user to type and press Enter
3. Returns whatever they typed as a **string**
4. Stores it in the variable

**Critical Detail**: `input()` ALWAYS returns a string, even if the user types a number!

## 🔥 Gotcha #10: Input Returns Strings

### 🤔 Predict: What will happen?

````python
# User types: 25
age = input("How old are you? ")
next_year = age + 1
print("Next year you'll be", next_year)
````

**Error!** `TypeError: can only concatenate str (not "int") to str`

### 🔍 Why?
Even though the user typed `25`, `input()` stored it as `"25"` (a string).

You can't do `"25" + 1` (can't add string and number).

**We need type conversion!**

# Part 5: Type Conversion

## Changing Data Types

Python provides functions to convert between types:

### The Conversion Functions

| Function | Purpose | Example |
|----------|---------|----------|
| `int()` | Convert to integer | `int("42")` → `42` |
| `float()` | Convert to float | `float("3.14")` → `3.14` |
| `str()` | Convert to string | `str(42)` → `"42"` |
| `bool()` | Convert to boolean | `bool(1)` → `True` |

## Converting Strings to Numbers

````python
# String to integer
age_text = "25"
age_number = int(age_text)
print("Type before:", type(age_text))
print("Type after:", type(age_number))
print("Can now do math:", age_number + 5)
````

````python
# String to float
price_text = "19.99"
price_number = float(price_text)
print("Type:", type(price_number))
print("Total for 3:", price_number * 3)
````

## Converting Numbers to Strings

````python
# Number to string (for concatenation)
score = 100
message = "Your score is " + str(score) + " points!"
print(message)
````

## 🔥 Gotcha #11: Conversion Can Fail

### 🤔 What happens if we try to convert invalid data?

````python
# This will cause an error
int("Hello")
````

````python
# This will also cause an error
int("3.14")
````

**ValueError**: `invalid literal for int() with base 10`

### What's Valid?
- ✅ `int("42")` - valid integer string
- ✅ `float("3.14")` - valid float string
- ✅ `float("42")` - can convert int string to float
- ❌ `int("3.14")` - can't directly convert float string to int
- ❌ `int("Hello")` - not a number at all

**Solution for "3.14" → int**: Convert to float first!
```python
int(float("3.14"))  # Returns 3
```

## Fixing Our Age Program

````python
# Get age from user and convert immediately
age = int(input("How old are you? "))
next_year = age + 1
print("Next year you'll be", next_year)
````

**Pattern**: Convert input immediately if you need a number:
```python
age = int(input("Age: "))           # For whole numbers
height = float(input("Height: "))   # For decimals
name = input("Name: ")              # Keep as string
```

# 🔥 Gotcha #12: Floating Point Precision (Bonus!)

### 🤔 Predict: What will this print?

````python
result = 0.1 + 0.2
print(result)
print(result == 0.3)
````

**Surprise!** `0.1 + 0.2 = 0.30000000000000004` (not exactly 0.3)

### Why?
Computers store decimal numbers in binary (0s and 1s). Some decimals can't be represented exactly, just like ⅓ can't be written exactly in decimal (0.333333...).

### What to Do?
For most programming, this tiny difference doesn't matter. But if you need exact decimal arithmetic (like money), use the `Decimal` module (we'll learn later!).

**Don't worry about this for now** - just know it exists!

# 💡 Debugging Tips

When your code doesn't work, follow this checklist:

## 1. Read the Error Message
- Start at the **bottom** (the actual error)
- Look at the **line number**
- Read what Python **expected**

## 2. Common Fixes
**TypeError: can only concatenate str**
→ Did you forget `str()` or use `,` instead of `+`?

**ValueError: invalid literal for int()**
→ Is the input actually a number? Try `type()` to check.

**NameError: name 'X' is not defined**
→ Did you spell the variable correctly? Is it defined above this line?

**SyntaxError: invalid syntax**
→ Missing quotes, parentheses, or colon? Check the line above too!

## 3. The Print Debug Method
When stuck, add print statements to see what's happening:

````python
age = input("Age: ")
print("DEBUG - age value:", age)
print("DEBUG - age type:", type(age))
age = int(age)
print("DEBUG - age after conversion:", age, type(age))
````
