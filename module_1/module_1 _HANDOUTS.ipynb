
1. **You write readable code** like `print("Hello")`.

- **Exact instructions**: It does only what you tell it.

# Part 2: Variables

A **variable** is a name that points to a value so you can use that value later in your program.

## Creating Your First Variable

````python
# Create a variable
age = 25
print(age)
````

### What Happened?
1. `age` is the **variable name** (the label)
2. `=` is the **assignment operator** ("point this label at...")
3. `25` is the **value** (what the label points to)
4. Python created a spot in memory, stored `25`, and labeled it `age`

## Variable Naming Rules

### Valid Variable Names:
- Must start with a letter or underscore: `age`, `_count`
- Can contain letters, numbers, underscores: `user_name_2`
- **Case-sensitive**: `age` ≠ `Age` ≠ `AGE`

### Invalid Variable Names:
- Can't start with a number: `2fast`
- Can't use special characters: `user-name`, `my.var`
- Can't use Python keywords: `print`, `if`, `for`

### Python Naming Convention (Snake Case):
Use lowercase with underscores:
- `user_name` — Python standard
- `userName` — JavaScript/Java style, avoid in Python
- `UserName` — reserved for class names

**Why does this matter?** Consistency makes code written by others (and your future self) readable.


## Multiple Variables

````python
# Create multiple variables
name = "Alice"
age = 25
city = "Stockholm"

print(name)
print(age)
print(city)
````

### Variables Can Change (They're "Variable" After All!)

````python
age = 25
print("Age is:", age)

# One year passes...
age = 26
print("Age is now:", age)
````

**What happened?** The label `age` now points to a different value in memory. The old value (`25`) gets forgotten.

## Data Types

Python needs to know what **kind** of data you're working with. This affects what you can do with it.

## The Four Core Types

### 1. `str` (String) - Text Data
- Anything wrapped in quotes
- Single quotes `'Hello'` or double quotes `"Hello"` (your choice!)
- Examples: `"Alice"`, `'Stockholm'`, `"42"`, `"True"`

### 2. `int` (Integer) - Whole Numbers
- Numbers without decimals
- Can be positive, negative, or zero
- Examples: `42`, `-10`, `0`, `1000`

### 3. `float` (Floating-Point) - Decimal Numbers
- Numbers with decimals
- Examples: `3.14`, `-0.5`, `2.0`
- **Note**: Even `2.0` is a float (has a decimal point)

### 4. `bool` (Boolean) - True or False
- Only two values: `True` or `False`
- **Case-sensitive**: `True` ✅, `true` ❌
- Used for decisions (more on this later!)


Look at these values — can you guess their types?

```python
"Hello"      # Type: ?
42           # Type: ?
3.14         # Type: ?
True         # Type: ?
"42"         # Type: ?
```

Let's check:

````python
# The type() function reveals the type
print(type("Hello"))
print(type(42))
print(type(3.14))
print(type(True))
print(type("42"))  # Notice the quotes!
````

## Why Do Types Matter?

Different types behave differently:

````python
# Numbers can be added
print(10 + 5)  # What will this print?
````

````python
# Strings can be "added" too (concatenation)
print("Hello" + " " + "World")  # What will this print?
````

````python
# But you can't mix types easily
print("I am " + 25 + " years old")  # What will happen?
````

### Gotcha #3: Can't Add Strings and Numbers

**Error!** `TypeError: can only concatenate str (not "int") to str`

### Why?
Python doesn't know if you want:
- `"I am " + "25"` (string concatenation) -> `"I am 25"`
- `"I am " + 25` (maybe you meant to add the number?)

So it refuses to guess!

### Three Solutions:

````python
age = 25

# Solution 1: Convert to string with str()
message = "I am " + str(age) + " years old"
print(message)

# Solution 2: Use comma separation (easier!)
print("I am", age, "years old")

# Solution 3: f-strings (we'll learn this later)
print(f"I am {age} years old")
````

## Type Checking: The `type()` Function

When in doubt, check!

````python
age = 25
name = "Alice"
height = 1.65
is_student = True

print("age is type:", type(age))
print("name is type:", type(name))
print("height is type:", type(height))
print("is_student is type:", type(is_student))
````

# Part 3: Operations

## Arithmetic Operations

Python can do all basic math operations and much more.

### Basic Operators


````python
# Addition
print("10 + 5 =", 10 + 5)

# Subtraction
print("10 - 5 =", 10 - 5)

# Multiplication
print("10 * 5 =", 10 * 5)

# Division
print("10 / 5 =", 10 / 5)
````

## More Powerful Operators

````python
# Exponentiation (power)
print("2 ** 3 =", 2 ** 3)  # 2 to the power of 3
print("10 ** 2 =", 10 ** 2)  # 10 squared

# Integer Division (floor division)
print("17 // 5 =", 17 // 5)  # Gives whole number only

# Modulo (remainder)
print("17 % 5 =", 17 % 5)  # What's left over?
````

### When is Modulo Useful?
- **Check if a number is even**: `number % 2 == 0`
- **Get last digit**: `number % 10`
- **Wrap around**: Think of a clock (13 % 12 = 1)

## Comparison Operators

Comparison operators ask questions about your data. They always return `True` or `False`.

### The Six Comparison Operators


````python
x = 10
y = 5

print("x == y:", x == y)  # Equal to
print("x != y:", x != y)  # Not equal to
print("x > y:", x > y)    # Greater than
print("x < y:", x < y)    # Less than
print("x >= y:", x >= y)  # Greater than or equal to
print("x <= y:", x <= y)  # Less than or equal to
````

## Gotcha #9: = vs ==

**This is one of the most common beginner mistakes!**

### `=` is Assignment ("becomes")

````python
x = 10  # x becomes 10
print(x)
````

### `==` is Comparison ("is equal to?")

````python
x == 10  # Is x equal to 10? Returns True or False
print(x == 10)
print(x == 5)
````

**Remember**:
- `=` → Makes a variable
- `==` → Asks a question

## Comparisons with Strings

````python
name1 = "Alice"
name2 = "alice"
name3 = "Alice"

print(name1 == name3)  # True
print(name1 == name2)  # False - case sensitive!
````

**Remember**: String comparisons are **case-sensitive**!
- `"Alice"` ≠ `"alice"`
- `"Python"` ≠ `"python"`

````python
# Strings also support >, <, >=, <= (alphabetical/lexicographic order)
print("apple" < "banana")    # True
print("zebra" > "yak")       # True
print("cat" >= "cat")        # True
print("ant" <= "Ant")        # False (uppercase comes before lowercase)
````

# Part 4: Input & Type Conversion

## Getting User Input

So far, all our variables have been hardcoded. Let's make our programs talk to users.

### The `input()` Function


````python
# Get user's name
name = input("What is your name? ")
print("Hello,", name, "!")
````

### How `input()` Works:
1. Displays the prompt (the text in quotes)
2. Waits for user to type and press Enter
3. Returns whatever they typed as a **string**
4. Stores it in the variable

**Critical Detail**: `input()` ALWAYS returns a string, even if the user types a number!

## Gotcha #10: Input Returns Strings

### Predict: What will happen?

````python
# User types: 25
age = input("How old are you? ")
next_year = age + 1
print("Next year you'll be", next_year)
````

**Error!** `TypeError: can only concatenate str (not "int") to str`

### 🔍 Why?
Even though the user typed `25`, `input()` stored it as `"25"` (a string).

You can't do `"25" + 1` (can't add string and number).

**We need type conversion!**

---

## Type Conversion

Python provides functions to convert between data types:

| Function | Purpose | Example |
|----------|---------|---------|
| `int()` | Convert to integer | `int("42")` → `42` |
| `float()` | Convert to float | `float("3.14")` → `3.14` |
| `str()` | Convert to string | `str(42)` → `"42"` |
| `bool()` | Convert to boolean | `bool(1)` → `True` |


## Converting Strings to Numbers

````python
# String to integer
age_text = "25"
age_number = int(age_text)
print("Type before:", type(age_text))
print("Type after:", type(age_number))
print("Can now do math:", age_number + 5)
````

````python
# String to float
price_text = "19.99"
price_number = float(price_text)
print("Type:", type(price_number))
print("Total for 3:", price_number * 3)
````

## Converting Numbers to Strings

````python
# Number to string (for concatenation)
score = 100
message = "Your score is " + str(score) + " points!"
print(message)
````

## Gotcha #11: Conversion Can Fail

### What happens if we try to convert invalid data?

````python
# This will cause an error
int("Hello")
````

````python
# This will also cause an error
int("3.14")
````

**ValueError**: `invalid literal for int() with base 10`

### What's Valid?
- ✅ `int("42")` - valid integer string
- ✅ `float("3.14")` - valid float string
- ✅ `float("42")` - can convert int string to float
- ❌ `int("3.14")` - can't directly convert float string to int
- ❌ `int("Hello")` - not a number at all

**Solution for "3.14" → int**: Convert to float first!
```python
int(float("3.14"))  # Returns 3
```

## Fixing Our Age Program

````python
# Get age from user and convert immediately
age = int(input("How old are you? "))
next_year = age + 1
print("Next year you'll be", next_year)
````

**Fun Fact**: `+=` is shorthand for adding and reassigning.

`age += 1` is the same as `age = age + 1`.

**Pattern**: Convert input immediately if you need a number:
```python
age = int(input("Age: "))           # For whole numbers
height = float(input("Height: "))   # For decimals
name = input("Name: ")              # Keep as string
```

## Troubleshooting & Getting Help

### Read the Error Message
Python error messages are your best friend. When something breaks:
1. Read the **last line** first — it states the error type and a short description
2. Note the **line number** — that's where Python got confused (the real problem may be one line above)
3. Common types and what they mean:
   - `SyntaxError` — Python can't parse the code; check for missing quotes, colons, or parentheses
   - `NameError` — a variable is used before it was defined, or is misspelled
   - `TypeError` — an operation is applied to the wrong type (e.g. adding a string and an integer)
   - `ValueError` — the type is right but the value is invalid (e.g. `int("hello")`)
   - `IndentationError` — inconsistent whitespace; check your indentation

### The Print Debug Method
When the error isn't obvious, add temporary `print()` calls to inspect what's happening:

```python
age = input("Age: ")
print("DEBUG - age value:", age, "| type:", type(age))
age = int(age)
print("DEBUG - age after conversion:", age)
```

Remove the debug lines once the issue is resolved.
