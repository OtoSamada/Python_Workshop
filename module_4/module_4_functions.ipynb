# Part 1: Built-in Functions - Python's Toolkit

Before writing our own functions, let's review useful built-in functions.

## Common Built-in Functions

````python
numbers = [3, 1, 4, 1, 5, 9, 2, 6]

# Basic operations
print("Length:", len(numbers))
print("Sum:", sum(numbers))
print("Min:", min(numbers))
print("Max:", max(numbers))

# Sorting (returns new list)
print("Sorted:", sorted(numbers))
print("Original:", numbers)  # Unchanged

# Reversing (returns iterator)
print("Reversed:", list(reversed(numbers)))
````

## Boolean Functions

`any()` and `all()` test whether items in a collection satisfy a condition.

> **Note**: The examples below use a **generator expression** — `condition for item in iterable` — which is syntactically similar to the list comprehensions covered in Part 5. For now, read them as "check this condition for each item."

````python
numbers = [1, 2, 3, 4, 5]

# any(condition for item in iterable) — True if AT LEAST ONE item satisfies the condition
print("Any positive?", any(n > 0 for n in numbers))   # True
print("Any > 10?",    any(n > 10 for n in numbers))   # False

# all(condition for item in iterable) — True if ALL items satisfy the condition
print("All positive?", all(n > 0 for n in numbers))   # True
print("All > 3?",      all(n > 3 for n in numbers))   # False
````

# Part 2: Function Basics

## Why Functions?

**Problem**: Repetitive code is hard to maintain

````python
# Without functions - repetitive!
name1 = "Alice"
print(f"Hello, {name1}!")
print(f"Welcome to Python, {name1}!")

name2 = "Bob"
print(f"Hello, {name2}!")
print(f"Welcome to Python, {name2}!")

# If we want to change the greeting, we have to update it everywhere!
````

**Solution**: Write it once, use it many times

````python
# With functions - DRY (Don't Repeat Yourself)
def greet(name):
    print(f"Hello, {name}!")
    print(f"Welcome to Python, {name}!")

greet("Alice")
greet("Bob")
````

## Anatomy of a Function

````python
def calculate_area(radius):
    """Calculate the area of a circle."""
    pi = 3.14159
    area = pi * radius ** 2
    return area

# Use the function
result = calculate_area(5)
print(f"Area: {result}")
````

**Components**:
1. **`def`** - Keyword to define a function
2. **`calculate_area`** - Function name (use descriptive names)
3. **`(radius)`** - Parameter(s)
4. **`:`** - Colon starts the function body
5. **Docstring** - Optional description (best practice!)
6. **Function body** - Indented code that runs when called
7. **`return`** - Sends a value back to the caller

## Parameters vs Arguments

These terms are often confused:

```python
def greet(name):        # 'name' is a PARAMETER
    print(f"Hello, {name}")

greet("Alice")          # "Alice" is an ARGUMENT
```

- **Parameter**: Variable in the function definition
- **Argument**: Actual value passed when calling the function

## Multiple Parameters

````python
def calculate_rectangle_area(length, width):
    """Calculate area of a rectangle."""
    return length * width

area = calculate_rectangle_area(10, 5)
print(f"Rectangle area: {area}")
````

````python
# Can use keyword arguments
area = calculate_rectangle_area(width=5, length=10)  # Order doesn't matter
print(f"Rectangle area: {area}")
````

## Return Values

Functions can send data back to the caller:

````python
def add(a, b):
    return a + b

result = add(5, 3)
print(result)  # 8
````

**Important**: `return` stops function execution immediately

````python
def check_positive(number):
    if number > 0:
        return "Positive"
    return "Not positive"  # Only reached if number <= 0

print(check_positive(5))
print(check_positive(-3))
````

## Gotcha #34: Forgetting the Return Statement

If you forget `return`, the function silently returns `None`:

````python
def add_wrong(a, b):
    result = a + b  # Forgot to return!

answer = add_wrong(5, 3)
print(answer)  # None
print(type(answer))  # <class 'NoneType'>
````

**Fix**: Add `return` statement

````python
def add_correct(a, b):
    result = a + b
    return result  # Now it works

answer = add_correct(5, 3)
print(answer)  # 8
````

## Multiple Return Values

Functions can return multiple values as a tuple:

````python
def get_min_max(numbers):
    """Return both minimum and maximum values."""
    return min(numbers), max(numbers)

# Unpack the returned tuple
minimum, maximum = get_min_max([3, 1, 4, 1, 5, 9])
print(f"Min: {minimum}, Max: {maximum}")
````

**Critical distinction**:

````python
# Definition - creates the function (doesn't execute it)
def greet(name):
    print(f"Hello, {name}!")

# Nothing printed yet!

# Call - executes the function
greet("Alice")  # Now it prints
````

# Part 3: Scope — Where Variables Live

## Local vs Global Scope

**Scope** is the region of your program where a particular variable name is visible and accessible.

Python has two main scopes:
- **Global scope** — variables defined at the top level of the file; accessible everywhere
- **Local scope** — variables defined inside a function; only accessible inside that function

This is intentional: functions are meant to be **self-contained**. They receive data through parameters and return data through `return`, rather than freely reading and writing shared state.

````python
# Global variable
global_var = "I'm global"

def my_function():
    # Local variable
    local_var = "I'm local"
    print(global_var)  # Can access global
    print(local_var)   # Can access local

my_function()
print(global_var)      # Can access global
# print(local_var)     # ERROR: local_var doesn't exist here
````

**Rule**: Variables created inside a function only exist inside that function.

## Visualizing Scope

````python
x = 10  # Global x

def modify():
    x = 20  # Local x (different variable!)
    print(f"Inside function: x = {x}")

modify()
print(f"Outside function: x = {x}")  # Global x unchanged
````

## Gotcha #35: Shadowing Global Variables

````python
count = 0

def increment():
    count = count + 1  # Error: can't read count before assignment
    return count

# increment()  # UnboundLocalError
````

**Why?** Python sees `count = ...` and treats `count` as local. But it tries to read `count` first, which doesn't exist locally yet.

**Solution 1**: Pass as parameter

````python
def increment(count):
    return count + 1

count = 0
count = increment(count)
print(count)  # 1
````

## Parameters Create Local Variables

````python
def calculate(x, y):
    # x and y are local variables
    result = x + y  # result is also local
    return result

answer = calculate(5, 3)
print(answer)
# print(x)      # Error: x doesn't exist here
# print(result) # Error: result doesn't exist here
````

## Best Practice: Avoid Global Variables

**Why?**
- Makes code harder to understand
- Functions should get data through parameters
- Functions should return data, not modify globals

**Good practice**:

````python
# Good: Everything explicit
def calculate_total(prices, tax_rate):
    subtotal = sum(prices)
    total = subtotal * (1 + tax_rate)
    return total

prices = [10, 20, 30]
tax_rate = 0.08
total = calculate_total(prices, tax_rate)
print(f"Total: ${total:.2f}")
````

# Part 4: Advanced Function Features

## Default Parameters

Parameters can have default values:

````python
def greet(name, greeting="Hello"):
    """Greet someone with a customizable greeting."""
    print(f"{greeting}, {name}!")

greet("Alice")              # Uses default: "Hello"
greet("Bob", "Hi")          # Overrides default
greet("Charlie", greeting="Hey")  # Keyword argument
````

**Rule**: Default parameters must come **after** required parameters

```python
def func(required, optional="default"):  # ✓ Correct
def func(optional="default", required):  # ✗ SyntaxError
```

## Gotcha #36: Mutable Default Arguments

````python
# DANGEROUS: Mutable default argument
def add_item(item, items=[]):
    items.append(item)
    return items

print(add_item("apple"))   # ['apple']
print(add_item("banana"))  # ['apple', 'banana'] - Unexpected!
````

**Problem**: Default list is created once and reused!

**Fix**: Use `None` as default

````python
def add_item_safe(item, items=None):
    if items is None:
        items = []  # Create new list each time
    items.append(item)
    return items

print(add_item_safe("apple"))   # ['apple']
print(add_item_safe("banana"))  # ['banana'] - Correct!
````

## *args: Variable Number of Arguments

Accept any number of positional arguments:

````python
def add_all(*numbers):
    """Add any number of arguments."""
    total = 0
    for num in numbers:
        total += num
    return total

print(add_all(1, 2, 3))           # 6
print(add_all(1, 2, 3, 4, 5))     # 15
print(add_all(10))                # 10
````

**How it works**: `*numbers` collects all arguments into a tuple

````python
def show_args(*args):
    print(f"Received: {args}")
    print(f"Type: {type(args)}")

show_args(1, 2, 3, 4)
````

## **kwargs: Keyword Arguments

Accept any number of keyword arguments:

````python
def describe_person(**info):
    """Print information about a person."""
    for key, value in info.items():
        print(f"{key}: {value}")

describe_person(name="Alice", age=25, city="Stockholm")
print()
describe_person(name="Bob", occupation="Engineer")
````

**How it works**: `**info` collects keyword arguments into a dictionary

````python
def show_kwargs(**kwargs):
    print(f"Received: {kwargs}")
    print(f"Type: {type(kwargs)}")

show_kwargs(name="Alice", age=25)
````

## Combining All Argument Types

**Order matters**: regular → *args → **kwargs

````python
def flexible_function(required, *args, optional="default", **kwargs):
    print(f"Required: {required}")
    print(f"Args: {args}")
    print(f"Optional: {optional}")
    print(f"Kwargs: {kwargs}")

flexible_function(
    "must_have",
    "extra1", "extra2",
    optional="custom",
    name="Alice",
    age=25
)
````

## Lambda Functions: Anonymous One-Liners

A **lambda function** is a small, nameless function defined in a single expression.

**Syntax**: `lambda parameters: expression`

```python
# Regular function
def square(x):
    return x ** 2

# Equivalent lambda
square = lambda x: x ** 2
```

Lambdas are most useful when you need a short function **in-place** — passed directly as an argument to another function — rather than defining a named function separately.

````python
# Basic lambda examples
square = lambda x: x ** 2
add = lambda x, y: x + y

print(square(5))   # 25
print(add(3, 4))   # 7

# The real power: passing a lambda directly as an argument
numbers = [3, -1, 4, -1, 5, -9, 2, -6]

# Sort by absolute value — lambda used inline
sorted_by_abs = sorted(numbers, key=lambda x: abs(x))
print("Sorted by absolute value:", sorted_by_abs)

# Sort a list of dictionaries by a field
students = [
    {"name": "Charlie", "grade": 88},
    {"name": "Alice",   "grade": 95},
    {"name": "Bob",     "grade": 72},
]
by_grade = sorted(students, key=lambda s: s["grade"], reverse=True)
for s in by_grade:
    print(f"{s['name']}: {s['grade']}")
````

**Lambda rules & limitations**:
- Can only contain a **single expression** — no `if/else` blocks, no loops, no `return`
- Great for short `key=` arguments in `sorted()`, `min()`, `max()`
- If the logic is more than a couple of tokens, use a regular `def` — it's more readable

```python
# Good use of lambda — short key function
sorted(words, key=lambda w: len(w))

# Bad use of lambda — just use def instead
process = lambda x: x ** 2 + x - 1  # Harder to read; give it a name!
```

# Part 5: List Comprehensions

## Pythonic Data Transformation

List comprehensions provide a concise way to create lists.

### Traditional Approach

````python
# Create list of squares
squares = []
for i in range(10):
    squares.append(i ** 2)

print(squares)
````

### List Comprehension

````python
# Same result, one line
squares = [i ** 2 for i in range(10)]
print(squares)
````

**Syntax**: `[expression for item in iterable]`

## List Comprehension with Conditions

````python
# Get only even numbers
evens = [i for i in range(20) if i % 2 == 0]
print(evens)

# Square only positive numbers
numbers = [-2, -1, 0, 1, 2, 3]
positive_squares = [n ** 2 for n in numbers if n > 0]
print(positive_squares)
````

**Syntax**: `[expression for item in iterable if condition]`

## Transforming Strings

````python
# Convert list of strings to uppercase
words = ["hello", "world", "python"]
upper_words = [word.upper() for word in words]
print(upper_words)

# Get lengths of each word
lengths = [len(word) for word in words]
print(lengths)

# Get first character of each word
first_chars = [word[0] for word in words]
print(first_chars)
````

## Nested List Comprehensions

````python
# Create multiplication table
table = [[i * j for j in range(1, 6)] for i in range(1, 6)]

for row in table:
    print(row)
````

**Note**: Nested comprehensions can be hard to read. Use regular loops if it's clearer.

## Dictionary and Set Comprehensions

The same idea extends to other collection types.

### Dictionary Comprehensions

**Syntax**: `{key_expr: value_expr for item in iterable}`

We'll also use the built-in `zip(a, b)` below — it pairs two iterables together, element by element.

````python
# Create a dictionary of squares
squares_dict = {i: i**2 for i in range(5)}
print(squares_dict)  # {0: 0, 1: 1, 2: 4, 3: 9, 4: 16}

# zip(a, b) pairs up two iterables element-by-element: zip([1,2,3], ['a','b','c']) → (1,'a'), (2,'b'), (3,'c')
keys   = ["a", "b", "c"]
values = [1, 2, 3]
mapping = {k: v for k, v in zip(keys, values)}
print(mapping)  # {'a': 1, 'b': 2, 'c': 3}

# zip() is also useful outside comprehensions
pairs = list(zip(keys, values))
print(pairs)  # [('a', 1), ('b', 2), ('c', 3)]
````

## Set Comprehensions

````python
# Create set (removes duplicates)
numbers = [1, 2, 2, 3, 3, 4, 5, 5]
unique_squares = {n**2 for n in numbers}
print(unique_squares)
````

# Part 6: Writing Good Functions — Best Practices

Building a function that *works* is step one. Writing a function that is easy for others (and future-you) to **read, trust, and maintain** is what separates beginner Python from professional Python.

The four practices below are industry standards — they apply to every language, not just Python.

## 1. Use Descriptive Names

````python
# Bad
def calc(x, y):
    return x * y

# Good
def calculate_rectangle_area(length, width):
    return length * width
````

## 2. Write Docstrings

````python
def calculate_average(numbers):
    """
    Calculate the arithmetic mean of a list of numbers.
    
    Args:
        numbers: List of numeric values
    
    Returns:
        Float representing the average
    """
    return sum(numbers) / len(numbers)

# Access docstring
help(calculate_average)
````

## 3. Keep Functions Focused

Each function should do **one thing** well.

````python
# Bad: One massive function doing everything
def process_student_data(raw_csv_line):
    # Validate format, clean name, convert grade, check threshold,
    # build output string, AND print — all in one function.
    # Changing any single step risks breaking the others.
    parts = raw_csv_line.split(",")
    name = parts[0].strip().title()
    grade = float(parts[1].strip())
    if grade < 0 or grade > 100:
        raise ValueError("Invalid grade")
    status = "Pass" if grade >= 50 else "Fail"
    print(f"{name}: {grade:.1f} ({status})")

# Good: Each function has a single, clear responsibility
def parse_student(raw_csv_line):
    """Parse 'name, grade' CSV line into (name, grade) tuple."""
    name, grade = raw_csv_line.split(",")
    return name.strip().title(), float(grade.strip())

def validate_grade(grade):
    """Raise ValueError if grade is outside 0–100."""
    if not (0 <= grade <= 100):
        raise ValueError(f"Grade {grade} is out of range")

def grade_status(grade):
    """Return 'Pass' or 'Fail' based on grade."""
    return "Pass" if grade >= 50 else "Fail"

# Main logic simply orchestrates the small pieces
name, grade = parse_student("alice, 87")
validate_grade(grade)
print(f"{name}: {grade:.1f} ({grade_status(grade)})")
````

## 4. Avoid Side Effects

A **side effect** is when a function modifies something outside its own local scope — like changing a list that was passed in, updating a global variable, or printing output the caller didn't ask for.

**Prefer functions that take inputs and return a result** — leave it to the caller to decide what to do with that result.

````python
# Side effect: silently mutates the caller's list
def add_item_bad(items, item):
    items.append(item)  # Changes the original list — caller may not expect this!

my_list = [1, 2, 3]
add_item_bad(my_list, 4)
print(my_list)  # [1, 2, 3, 4] — the original was modified without warning!

# No side effect: return a new list, leave the original untouched
def add_item_good(items, item):
    return items + [item]  # Creates and returns a brand-new list

my_list = [1, 2, 3]
new_list = add_item_good(my_list, 4)
print(my_list)   # [1, 2, 3]    — unchanged
print(new_list)  # [1, 2, 3, 4] — caller decides what to do with the result
````

| # | Name | Problem | Fix |
|---|------|---------|-----|
| 34 | Forgetting `return` | Function body computes a result but never returns it — caller receives `None` | Add `return result` as the last line |
