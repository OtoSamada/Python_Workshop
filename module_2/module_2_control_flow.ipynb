# Part 1: Conditionals — Making Decisions

## The `if` Statement

Until now, every line of our programs ran in sequence. Real programs need to respond to different situations:
- If it's raining, bring an umbrella
- If the password is correct, log in
- If age >= 18, allow entry

Python's `if` statement lets us express exactly this kind of logic.


## Your First If Statement

````python
age = 20

if age >= 18:
    print("You are an adult!")
    print("You can vote!")

print("Program continues...")
````

### Anatomy of an `if` Statement

```python
if condition:
    code_to_run
    more_code
```

1. **`if`** — the keyword that starts the decision
2. **`condition`** — an expression that evaluates to `True` or `False`
3. **`:`** — required; tells Python that a block follows
4. **Indentation** — 4 spaces (or 1 tab); defines what code belongs inside the `if`


## Understanding Indentation (Critical!)

In most languages, indentation is a style choice. **In Python, indentation is syntax.**

```
if it's raining:
    grab umbrella          ← inside the if block (indented)
    put on raincoat        ← still inside (indented)
lock the door              ← outside the block — always runs
```

Indentation tells Python which lines of code belong together.


````python
# Let's see this in action
temperature = 15

if temperature < 20:
    print("It's cold!")           # Runs only if temperature < 20
    print("Wear a jacket!")       # Also only if temperature < 20

print("Have a nice day!")         # ALWAYS runs (not indented)
````

**Note**: Use your editor's Tab key for indentation. Never mix tabs and spaces in the same file.


## Conditions Evaluate to True or False

Let's see what's actually happening:

````python
age = 20

# The condition evaluates to a boolean
result = age >= 18
print("Condition result:", result)
print("Type:", type(result))

# The if statement uses this True/False value
if result:
    print("This runs because result is True!")
````

**Key insight**: Conditions are just expressions that evaluate to `True` or `False`. There is no magic — the `if` statement simply checks the boolean result.

## `if/elif/else`: Multiple Conditions

When you need to test more than two possibilities, use `elif` (short for "else if").


````python
score = 85

if score >= 90:
    print("Grade: A")
elif score >= 80:
    print("Grade: B")
elif score >= 70:
    print("Grade: C")
elif score >= 60:
    print("Grade: D")
else:
    print("Grade: F")
````

### How it works:
1. Check the `if` condition
2. If `False`, check the first `elif`
3. If still `False`, check the next `elif`
4. Continue down the chain
5. If all conditions are `False`, run the `else` block

**As soon as one condition is `True`, that block runs and all remaining conditions are skipped.**


## Gotcha #15: Order Matters

**What is wrong with this grading logic?**


````python
score = 95

if score >= 60:
    print("Grade: D")
elif score >= 70:
    print("Grade: C")
elif score >= 80:
    print("Grade: B")
elif score >= 90:
    print("Grade: A")
````

**Problem**: A score of 95 satisfies `score >= 60`, so it prints "Grade: D" and never evaluates the remaining conditions.

**Fix**: Order conditions from most specific (highest threshold) to least specific.

## Logical Operators: Combining Conditions

Often you need to check more than one condition at once:
- Access requires being **tall enough AND having a ticket**
- A discount applies to **students OR seniors**
- Access is denied to anyone **NOT on the guest list**

Python provides three logical operators: `and`, `or`, `not`.


## The `and` Operator


````python
# Movie theater: need to be 13+ AND have a ticket
age = 15
has_ticket = True

if age >= 13 and has_ticket:
    print("Welcome! Enjoy the movie!")
else:
    print("Sorry, you can't enter.")
````

**Truth table for `and`** — both conditions must be `True`:

| A     | B     | A and B |
|-------|-------|---------|
| True  | True  | True    |
| True  | False | False   |
| False | True  | False   |
| False | False | False   |


## The `or` Operator


````python
# Discount: for students OR seniors
is_student = False
is_senior = True

if is_student or is_senior:
    print("You get a 20% discount!")
else:
    print("Regular price.")
````

**Truth table for `or`** — at least one condition must be `True`:

| A     | B     | A or B |
|-------|-------|--------|
| True  | True  | True   |
| True  | False | True   |
| False | True  | True   |
| False | False | False  |


## The `not` Operator


````python
# Access control: deny if NOT on guest list
on_guest_list = False

if not on_guest_list:
    print("Access denied.")
else:
    print("Welcome!")
````

**Truth table for `not`** — inverts the boolean:

| A     | not A |
|-------|-------|
| True  | False |
| False | True  |


## Combining Operators


````python
# Complex condition: Weekend plans
is_weekend = True
weather = "sunny"
has_money = True

# Go to beach if: weekend AND (sunny weather OR have money for indoor backup)
if is_weekend and (weather == "sunny" or has_money):
    print("Let's go to the beach!")
else:
    print("Stay home.")
````

Use parentheses `()` to group and clarify complex conditions, just as in arithmetic.

# Part 2: Loops — Repeating Actions

## Why Loops?

Computers excel at repetitive tasks: counting, scanning data, processing images pixel by pixel. Writing the same code 100 times is not the solution — loops are.

Python has two types of loops:
1. **`while` loops** — repeat as long as a condition is `True`
2. **`for` loops** — repeat once for each item in a sequence


## `while` Loops

A `while` loop repeats its body as long as the condition remains `True`.


````python
# Count to 5
count = 1

while count <= 5:
    print(count)
    count = count + 1  # Don't forget to update!

print("Done counting!")
````

### Anatomy of a `while` Loop

```python
initialization          # Set the loop variable before the loop
while condition:        # Check whether to continue
    code_to_repeat      # Do the work
    update_variable     # Move toward the exit condition
```

The pattern is always: **Initialize → Check → Execute → Update → Check → ...**


## ⚠️ Avoiding Infinite Loops

If a loop's condition never becomes `False`, the program runs forever. This is a **common mistake** — not a disaster.

**To stop an infinite loop:**
- In Jupyter: click the **Stop (⏹)** button in the toolbar, or go to **Kernel → Interrupt**
- In the terminal: press `Ctrl+C`

The commented-out code below demonstrates what an infinite loop looks like. Study it — don't run it.


````python
# WARNING: This is infinite! Stop it with Ctrl+C or Stop button
# Uncomment to see (then immediately stop it!)

# count = 1
# while count <= 5:
#     print(count)
#     # Oops! Forgot to update count!
````

The condition `count <= 5` is always `True` because `count` is never updated inside the loop.


## Gotcha #16: Forgetting to Update the Loop Variable


````python
# This is the most common loop mistake!
count = 1
while count <= 3:
    print("Counting:", count)
    # Missing: count = count + 1
    # This will print "Counting: 1" forever!
````

**Fix**: Update the loop variable inside the loop.

**Shorthand**: `count += 1` is equivalent to `count = count + 1`.


````python
# Correct version
count = 1
while count <= 3:
    print("Counting:", count)
    count += 1  # Now it will stop!
````

## Demo: Countdown


````python
# Count down from 10 to 1
count = 10

while count > 0:
    print(count)
    count -= 1  # Same as count = count - 1

print("Blast off! 🚀")
````

## `for` Loops: Iterating over a Sequence

A `for` loop repeats once for each item in a sequence. The loop variable is automatically updated each iteration.


````python
# Print each number
for number in [1, 2, 3, 4, 5]:
    print(number)
````

### How it works:
1. Assign the first item from the list to `number`
2. Execute the indented body
3. Assign the next item to `number`
4. Execute the body again
5. Repeat until the list is exhausted

The loop variable updates automatically — no manual increment needed.


## The `range()` Function

Rather than writing out a list like `[0, 1, 2, 3, 4]`, use `range()` to generate number sequences.


````python
# range(5) generates: 0, 1, 2, 3, 4
for i in range(5):
    print(i)
````

### Gotcha #17: `range()` is Exclusive at the End

`range(5)` produces `0, 1, 2, 3, 4` — it stops *before* reaching 5.

To include a specific upper bound, pass `n + 1`: `range(1, 6)` gives `1, 2, 3, 4, 5`.


## The Three Forms of `range()`


````python
# range(stop) - from 0 to stop-1
print("range(5):")
for i in range(5):
    print(i, end=" ")
print()  # New line

# range(start, stop) - from start to stop-1
print("\nrange(1, 6):")
for i in range(1, 6):
    print(i, end=" ")
print()

# range(start, stop, step) - custom increment
print("\nrange(0, 10, 2):")
for i in range(0, 10, 2):
    print(i, end=" ")
print()
````

| Form | Example | Result |
|------|---------|--------|
| `range(stop)` | `range(5)` | 0, 1, 2, 3, 4 |
| `range(start, stop)` | `range(1, 6)` | 1, 2, 3, 4, 5 |
| `range(start, stop, step)` | `range(0, 10, 2)` | 0, 2, 4, 6, 8 |


## Common `range()` Patterns


````python
# Count 1 to 10 (inclusive)
for i in range(1, 11):  # Need 11 because it stops before!
    print(i, end=" ")
print()

# Count backwards (countdown)
for i in range(10, 0, -1):
    print(i, end=" ")
print()
````

## `while` vs `for`: Choosing the Right Loop

**Use `for` when:**
- You know how many iterations are needed upfront
- You are iterating over a collection or sequence
- Examples: "Repeat 10 times", "Process each item in a list"

**Use `while` when:**
- The number of iterations is not known in advance
- You are repeating until something happens
- Examples: "Keep prompting until valid input is received", "Run until user quits"

**General guidance**: When in doubt, prefer `for` — it is less prone to infinite loops.


## `while` and `for` Side by Side


````python
# Same task, two ways:
# Print numbers 1 to 5

print("With while:")
count = 1
while count <= 5:
    print(count)
    count += 1

print("\nWith for:")
for count in range(1, 6):
    print(count)
````

The `for` version is more concise and removes the need to manage the loop variable manually.

## Loop Control: `break` and `continue`

Two statements give you finer control over loop execution:
- **`break`** — exit the loop immediately
- **`continue`** — skip the rest of the current iteration and move to the next


### `break`: Exit Early


````python
# Search for a number
target = 7

for number in range(1, 11):
    print(f"Checking {number}...")
    if number == target:
        print(f"Found {target}!")
        break  # Exit the loop immediately

print("Search complete.")
````

Once the target was found, `break` terminated the loop immediately. This is the standard pattern for stopping a search early.


### `continue`: Skip to the Next Iteration


````python
# Print only odd numbers
for number in range(1, 11):
    if number % 2 == 0:  # If even
        continue  # Skip to next iteration
    print(number)  # Only prints odd numbers
````

When `number` is even, `continue` skips the `print` statement and moves directly to the next iteration.


## Gotcha #18: `break` vs `continue`

Both affect loop execution, but in different ways:

```
for item in items:
    if condition1:
        break      ← exits the entire loop
    if condition2:
        continue   ← skips to the next iteration
    # code below here is skipped when continue fires
```

A quick way to remember: `break` is a full stop; `continue` is "next, please."

## Nested Loops

You can place a loop inside another loop. The inner loop runs to completion on every single iteration of the outer loop.

Common uses:
- Generating grids or 2D patterns
- Comparing every item against every other item
- Building multiplication tables


````python
# Print a square of stars
for row in range(5):      # Outer loop: 5 rows
    for col in range(5):  # Inner loop: 5 stars per row
        print("*", end="")
    print()  # New line after each row
````

For each iteration of the outer loop, the inner loop runs completely. Think of the outer loop as rows and the inner loop as columns.


## Demo: Multiplication Table with Nested Loops


````python
# Create a multiplication table (1-5)
for i in range(1, 6):
    for j in range(1, 6):
        result = i * j
        print(f"{i}x{j}={result:2}", end="  ")  # :2 pads to 2 spaces
    print()  # New line after each row
````

# Part 3: Mini-Project — Number Guessing Game

## Overview

We will build a complete interactive game that combines everything covered in this module:
- Conditionals (`if/elif/else`)
- `while` loops
- User input
- Random number generation

**The game:**
1. The computer picks a random number between 1 and 100
2. The player guesses
3. The computer responds with "too high", "too low", or "correct!"
4. The player has 7 attempts to win


## Phase 1: Core Game Logic


````python
import random

# Generate random number
secret_number = random.randint(1, 100)
max_attempts = 7
attempts = 0

print("Welcome to the Number Guessing Game!")
print(f"I'm thinking of a number between 1 and 100.")
print(f"You have {max_attempts} attempts to guess it!\n")

# Game loop
while attempts < max_attempts:
    # Get guess
    guess = int(input(f"Attempt {attempts + 1}/{max_attempts} - Enter your guess: "))
    attempts += 1
    
    # Check guess
    if guess == secret_number:
        print(f"🎉 Congratulations! You guessed it in {attempts} attempts!")
        break
    elif guess < secret_number:
        print("📈 Too low! Try higher.")
    else:
        print("📉 Too high! Try lower.")
    
    # Game over?
    if attempts == max_attempts:
        print(f"\n😢 Game Over! The number was {secret_number}.")

print("\nThanks for playing!")
````

## Phase 2: Proximity Hints


````python
import random

secret_number = random.randint(1, 100)
max_attempts = 7
attempts = 0

print("Welcome to the Number Guessing Game!")
print(f"I'm thinking of a number between 1 and 100.")
print(f"You have {max_attempts} attempts to guess it!\n")

while attempts < max_attempts:
    guess = int(input(f"Attempt {attempts + 1}/{max_attempts} - Enter your guess: "))
    attempts += 1
    
    if guess == secret_number:
        print(f"🎉 Congratulations! You guessed it in {attempts} attempts!")
        break
    
    # Calculate difference
    difference = abs(guess - secret_number)
    
    # Give hints based on how close
    if difference <= 5:
        print("🔥 Very close!")
    elif difference <= 10:
        print("♨️ Close!")
    elif difference <= 20:
        print("🌡️ Getting warmer...")
    else:
        print("🧊 Cold!")
    
    # Direction hint
    if guess < secret_number:
        print("📈 Go higher!")
    else:
        print("📉 Go lower!")
    
    if attempts == max_attempts:
        print(f"\n😢 Game Over! The number was {secret_number}.")

print("\nThanks for playing!")
````

### New Concepts in Phase 2

**`abs()` — Absolute value**
```python
difference = abs(guess - secret_number)
```
`abs()` returns the magnitude of a number regardless of sign: `abs(-10) == 10`. This lets us measure how far the guess is from the target without caring about direction.

**Two independent decision trees**

First we check proximity (`difference <= 5`, etc.), then we check direction (`guess < secret_number`). These are separate `if/elif` chains, not nested — each runs unconditionally.


## Phase 3: Input Validation


````python
import random

secret_number = random.randint(1, 100)
max_attempts = 7
attempts = 0

print("Welcome to the Number Guessing Game!")
print(f"I'm thinking of a number between 1 and 100.")
print(f"You have {max_attempts} attempts to guess it!\n")

while attempts < max_attempts:
    # Input validation loop
    while True:
        try:
            guess = int(input(f"Attempt {attempts + 1}/{max_attempts} - Enter your guess (1-100): "))
            
            # Check if in valid range
            if guess < 1 or guess > 100:
                print("⚠️ Please enter a number between 1 and 100!")
                continue
            
            break  # Valid input, exit validation loop
            
        except ValueError:
            print("⚠️ That's not a valid number! Please enter a number.")
    
    attempts += 1
    
    if guess == secret_number:
        print(f"🎉 Congratulations! You guessed it in {attempts} attempts!")
        break
    
    difference = abs(guess - secret_number)
    
    if difference <= 5:
        print("🔥 Very close!")
    elif difference <= 10:
        print("♨️ Close!")
    elif difference <= 20:
        print("🌡️ Getting warmer...")
    else:
        print("🧊 Cold!")
    
    if guess < secret_number:
        print("📈 Go higher!")
    else:
        print("📉 Go lower!")
    
    if attempts == max_attempts:
        print(f"\n😢 Game Over! The number was {secret_number}.")

print("\nThanks for playing!")
````

### New Concepts in Phase 3

**Error handling with `try/except`**
```python
try:
    guess = int(input("Enter number: "))
except ValueError:
    print("Not a valid number!")
```
The `try` block runs code that might raise an exception. If `int()` fails (e.g. the user types `"abc"`), Python falls into the `except` block instead of crashing.

**The validation loop pattern**
```python
while True:       # Loop indefinitely
    # get input
    # validate
    if valid:
        break     # Exit only when input is acceptable
```
`while True` combined with a conditional `break` is the standard Python idiom for collecting valid user input.


### Gotchas in Phase 3

**Gotcha #21: `break` in Nested Loops**
```python
while attempts < max_attempts:     # outer: game loop
    while True:                    # inner: validation loop
        # get valid input
        break  # only exits the INNER loop!

    attempts += 1  # this runs after validation, as intended
```
`break` always exits only the **innermost** loop. Make sure statements like `attempts += 1` are positioned at the correct indentation level.

**Gotcha #22: Forgetting to `break` out of `while True`**
```python
while True:
    # get input
    # forgot to break on valid input → infinite loop
```
Every `while True` loop must have a `break` reachable on a valid code path.

# Module 2 Summary

## Conditionals
- `if/elif/else` — branch based on conditions
- Indentation defines code blocks (it is syntax, not style)
- Conditions evaluate to `True` or `False`
- Logical operators: `and`, `or`, `not`

## Loops
- `while` — repeat while a condition holds; always update the loop variable
- `for` — iterate over a sequence; the variable updates automatically
- `range()` — generate integer sequences; 0-indexed, exclusive upper bound
- `break` — exit the loop immediately
- `continue` — skip to the next iteration
- Nested loops — inner loop runs fully for each outer iteration
