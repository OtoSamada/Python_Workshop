# Part 1: Why Object-Oriented Programming?

## The Problem: Data and Functions Are Separate

Imagine managing information about a bank account:

````python
# Without OOP - using dictionary and separate functions
account = {
    "holder": "Alice",
    "balance": 1000
}

def deposit(account, amount):
    account["balance"] += amount

def withdraw(account, amount):
    if account["balance"] >= amount:
        account["balance"] -= amount
    else:
        print("Insufficient funds")

# Usage
deposit(account, 500)
withdraw(account, 200)
print(account)
````

**Problems**:
- Data and functions are separate
- No connection between them except by convention
- Easy to make mistakes (wrong function, wrong data)
- Namespace pollution (many global functions)

## The Solution: Classes Bundle Data and Behavior

````python
# With OOP - everything related is together
class BankAccount:
    def __init__(self, holder, balance):
        self.holder = holder
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
    
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
        else:
            print("Insufficient funds")

# Usage - much cleaner!
account = BankAccount("Alice", 1000)
account.deposit(500)
account.withdraw(200)
print(f"{account.holder}: ${account.balance}")
````

**Benefits**:
- Data and behavior bundled together (**encapsulation**)
- Clearer, more intuitive syntax
- Related functionality is organized
- Easier to maintain and extend

## When to Use a Class

A useful rule of thumb: **if a dictionary plus a set of functions feels messy, a class will be cleaner**.

**Good fit for a class**:
- You need to create **many instances** of the same concept (users, products, game characters)
- A piece of data naturally comes bundled with **methods that only make sense on that data**
- You want to expose a clean interface while hiding internal details (**encapsulation**)

**Examples from real code**:
- `User` — stores name/email, exposes `.login()`, `.logout()`, `.reset_password()`
- `Product` — stores name/price/stock, exposes `.apply_discount()`, `.restock()`
- `BankAccount` — stores holder/balance, exposes `.deposit()`, `.withdraw()`

# Part 2: Class Basics

## Classes vs Objects

**Class**: Blueprint or template
- Defines what attributes and methods an object will have
- Like a cookie cutter

**Object (Instance)**: Concrete realization of the class
- Created from the class blueprint
- Like the actual cookie

```
Class Dog               Objects

Blueprint        →      dog1 (Buddy, 3 years)
(cookie cutter)  →      dog2 (Max, 5 years)
                 →      dog3 (Luna, 2 years)
```

## Creating Your First Class

````python
class Dog:
    """A simple Dog class."""
    
    def __init__(self, name, age):
        """Initialize a new dog."""
        self.name = name
        self.age = age
    
    def bark(self):
        """Make the dog bark."""
        print(f"{self.name} says Woof!")
    
    def birthday(self):
        """Celebrate the dog's birthday."""
        self.age += 1
        print(f"{self.name} is now {self.age} years old!")
````

## Anatomy of a Class

```python
class Dog:                          # ① Class definition — capitalise by convention
    def __init__(self, name, age):  # ② Constructor — runs automatically on creation
        self.name = name            # ③ Instance attribute — stores data on this object
        self.age = age              # ③ Instance attribute

    def bark(self):                 # ④ Instance method — behaviour, takes self
        print(f"{self.name} says Woof!")
```

| # | Component | What it does |
|---|-----------|-------------|
| ① | `class Dog:` | Declares the class; class names use `UpperCamelCase` |
| ② | `__init__` | Constructor — Python calls this automatically when you write `Dog(...)` |
| ③ | `self.name` | Instance attribute — every object gets its own copy |
| ④ | `bark(self)` | Instance method — a function attached to the object; `self` is always the first parameter |

**`self`** is how an object refers to itself. Every instance method must accept it as its first parameter, but you never pass it manually — Python inserts it for you.

## Creating Objects (Instances)

````python
# Create dog objects
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

# Access attributes
print(f"Dog 1: {dog1.name}, {dog1.age} years old")
print(f"Dog 2: {dog2.name}, {dog2.age} years old")

# Call methods
dog1.bark()
dog2.bark()

# Modify attributes
dog1.birthday()
````

**What happens when you create an object**:
1. Python creates a new object in memory
2. Calls `__init__` with the arguments
3. `self` refers to the newly created object
4. Attributes are attached to `self`
5. Returns the completed object

## Understanding `self`

Every time you call a method, Python automatically passes the object as the first argument. `self` is just the conventional name for that argument.

```python
dog1.bark()
# Python translates this to:
Dog.bark(dog1)  # dog1 is passed in as self
```

That is why every instance method must have `self` as its first parameter — Python needs somewhere to put the reference to the object being acted on.

> **Mental model**: think of `self` as the method's way of knowing *which* dog is barking. Without it, the method would have no way to access `self.name`.

# Part 3: Making Classes Useful

## The Problem with Printing Objects

````python
dog = Dog("Buddy", 3)
print(dog)  # <__main__.Dog object at 0x...> - Not helpful!
````

**Problem**: Default representation is just memory address.

**Solution**: Implement `__str__` method

## String Representation: `__str__`

`__str__` defines how an object looks when printed.

````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def __str__(self):
        return f"Dog(name={self.name}, age={self.age})"
    
    def bark(self):
        print(f"{self.name} says Woof!")

dog = Dog("Buddy", 3)
print(dog)  # Dog(name=Buddy, age=3) - Much better!
````

**`__str__` vs `__repr__`** — Python has two string-conversion methods:

| Method | Used by | Purpose |
|--------|---------|---------|
| `__str__` | `print()`, `str()`, f-strings | Human-readable output |
| `__repr__` | interactive REPL, `repr()`, debuggers | Unambiguous developer representation |

**Recommended practice**: always define `__repr__`; only add `__str__` when you want a different, friendlier display. A common pattern is to make `__repr__` return something that looks like the constructor call:

```python
def __repr__(self):
    return f"Dog(name={self.name!r}, age={self.age!r})"
    # → Dog(name='Buddy', age=3)
```

For beginners, defining `__str__` is sufficient — just be aware that `__repr__` exists and serves the debugging workflow.

## Practical Example: BankAccount with `__str__`

> **Note**: You'll see `BankAccount` redefined several times throughout this module. Each version deliberately builds on the previous one — adding `__str__`, then validation, then helper methods, then class attributes. In a real project these would all be a single class definition; here we grow it incrementally so each new concept is easy to isolate and study.

````python
class BankAccount:
    def __init__(self, holder, balance=0):
        self.holder = holder
        self.balance = balance
    
    def __str__(self):
        return f"BankAccount({self.holder}: ${self.balance:.2f})"
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount:.2f}")
        else:
            print("Deposit amount must be positive")
    
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        elif amount <= 0:
            print("Withdrawal amount must be positive")
        else:
            self.balance -= amount
            print(f"Withdrew ${amount:.2f}")
    
    def get_balance(self):
        return self.balance
````

````python
# Use the BankAccount class
account = BankAccount("Alice", 1000)
print(account)

account.deposit(500)
print(account)

account.withdraw(200)
print(account)

account.withdraw(2000)  # Try to withdraw too much
print(account)
````

## Methods Can Call Other Methods

````python
class BankAccount:
    def __init__(self, holder, balance=0):
        self.holder = holder
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
    
    def withdraw(self, amount):
        if self.has_sufficient_funds(amount):  # Call another method
            self.balance -= amount
        else:
            print("Insufficient funds")
    
    def has_sufficient_funds(self, amount):
        """Helper method."""
        return self.balance >= amount
````

# Part 4: Class vs Instance Attributes

## Instance Attributes (What We've Been Using)

Each object has its own copy:

````python
class Dog:
    def __init__(self, name):
        self.name = name  # Instance attribute

dog1 = Dog("Buddy")
dog2 = Dog("Max")

print(dog1.name)  # Buddy
print(dog2.name)  # Max - different values
````

## Class Attributes (Shared by All Instances)

Defined outside `__init__`, shared by all objects:

````python
class Dog:
    species = "Canis familiaris"  # Class attribute
    
    def __init__(self, name):
        self.name = name  # Instance attribute

dog1 = Dog("Buddy")
dog2 = Dog("Max")

print(dog1.species)  # Canis familiaris
print(dog2.species)  # Canis familiaris - same value
print(Dog.species)   # Can access via class too
````

**Use class attributes for**:
- Constants shared by all instances
- Default values
- Counters (tracking number of instances)

## Class Methods

Alongside instance methods (which receive `self`), Python supports **class methods** — methods that receive the **class itself** as their first argument, conventionally named `cls`.

You mark a class method with the `@classmethod` decorator (the `@` line immediately before `def`).

```python
class MyClass:
    count = 0  # class attribute

    @classmethod
    def get_count(cls):   # cls is the class, not an instance
        return cls.count
```

Class methods are most useful for:
- **Factory methods** — alternate constructors: `MyClass.from_string("...")`
- **Accessing or modifying class attributes** — like a shared counter

> **Decorator note**: a decorator (`@something`) is a shorthand that wraps a function/method with extra behaviour. You don't need to understand decorators deeply right now — just know that `@classmethod` means "this method belongs to the class, not to instances".

## Practical Example: Counting Instances

````python
class BankAccount:
    total_accounts = 0  # Class attribute
    
    def __init__(self, holder, balance=0):
        self.holder = holder
        self.balance = balance
        BankAccount.total_accounts += 1  # Increment class attribute
    
    @classmethod
    def get_total_accounts(cls):
        return cls.total_accounts

# Create accounts
acc1 = BankAccount("Alice")
acc2 = BankAccount("Bob")
acc3 = BankAccount("Charlie")

print(f"Total accounts: {BankAccount.get_total_accounts()}")
````

## Gotcha #41: Confusing Class and Instance Attributes

When you write `self.species = "something"` inside a method, Python creates a **new instance attribute** that shadows the class attribute — it does not modify the shared class attribute. Other instances continue to see the original class-level value.

````python
class Dog:
    species = "Canis familiaris"  # class attribute (shared)

dog1 = Dog()
dog2 = Dog()

# This looks like you're changing the class attribute, but you're not!
dog1.species = "Wolf"  # Creates an instance attribute on dog1 only

print(dog1.species)   # "Wolf"          — dog1's own shadow copy
print(dog2.species)   # "Canis familiaris" — class attribute unchanged
print(Dog.species)    # "Canis familiaris" — also unchanged

# To genuinely change the class attribute, go through the class:
Dog.species = "Domestic dog"
print(dog2.species)   # "Domestic dog" — now reflects the updated class attribute
print(dog1.species)   # "Wolf"         — dog1's shadow copy still overrides for dog1
````

# Part 6: Inheritance - Code Reuse

## The Problem: Repetitive Code

Imagine creating multiple similar classes:

````python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def eat(self):
        print(f"{self.name} is eating")

class Cat:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def eat(self):
        print(f"{self.name} is eating")

# Lots of duplicate code!
````

## Solution: Inheritance

Create a parent class with shared functionality:

````python
class Animal:
    """Parent class with shared functionality."""
    
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def eat(self):
        print(f"{self.name} is eating")
    
    def sleep(self):
        print(f"{self.name} is sleeping")

class Dog(Animal):  # Dog inherits from Animal
    """Child class with specific behavior."""
    
    def bark(self):
        print(f"{self.name} says Woof!")

class Cat(Animal):  # Cat inherits from Animal
    """Child class with specific behavior."""
    
    def meow(self):
        print(f"{self.name} says Meow!")
````

````python
# Create instances
dog = Dog("Buddy", 3)
cat = Cat("Whiskers", 2)

# Inherited methods work
dog.eat()    # From Animal
dog.sleep()  # From Animal
dog.bark()   # From Dog

cat.eat()    # From Animal
cat.sleep()  # From Animal
cat.meow()   # From Cat
````

**The IS-A test**: before making a child class, ask "is this genuinely a *type of* the parent?" A `Dog` **is an** `Animal` — inheritance makes sense. A `Car` **is not an** `Engine` — `Car` *has an* `Engine`, so use composition (store it as an attribute).

## Overriding Methods and Polymorphism

Child classes can **override** (replace) any method defined in the parent by defining a method with the same name.

This enables **polymorphism** — the ability to call the same method name on different object types and get the right behaviour each time. The for-loop below shows this: every object is an `Animal`, but calling `make_sound()` on each one triggers the *correct* version automatically.

````python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def make_sound(self):
        print("Some generic animal sound")

class Dog(Animal):
    def make_sound(self):  # Override parent method
        print(f"{self.name} says Woof!")

class Cat(Animal):
    def make_sound(self):  # Override parent method
        print(f"{self.name} says Meow!")

# Polymorphism in action
animals = [Dog("Buddy"), Cat("Whiskers"), Dog("Max")]

for animal in animals:
    animal.make_sound()  # Correct method called for each type
````

**Polymorphism** ("many forms") is the reason you can write `animal.make_sound()` in a loop and not care whether each item is a `Dog` or a `Cat`. The correct method is chosen automatically at runtime based on the actual type of the object. This is one of OOP's most powerful ideas — it lets you write **generic code** that works across an entire family of types.

## Extending the Parent's `__init__` with `super()`

When a child class needs **extra attributes** beyond what the parent stores, it can define its own `__init__`. But it still needs to initialise the parent's attributes too — that's what `super().__init__(...)` does. 

Think of it as saying: *"Do everything the parent would normally do when creating an instance, then I'll add my own things on top."*

````python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Dog(Animal):
    def __init__(self, name, age, breed):
        super().__init__(name, age)  # Call parent's __init__
        self.breed = breed  # Add dog-specific attribute
    
    def __str__(self):
        return f"Dog({self.name}, {self.age}yo, {self.breed})"

dog = Dog("Buddy", 3, "Golden Retriever")
print(dog)
````

**`super()`** returns a proxy object that lets you call methods on the *parent* class. Using it rather than hard-coding the parent class name (e.g. `Animal.__init__(self, ...)`) is better practice: it keeps working correctly even if you later change the inheritance chain.

## Gotcha #40: Forgetting to Call `super().__init__()`

If a child class defines its own `__init__` but never calls the parent's, the parent attributes are never created — and any method that relies on them will raise an `AttributeError`.

````python
class Animal:
    def __init__(self, name):
        self.name = name

# Bad: parent __init__ is never called
class Dog(Animal):
    def __init__(self, name, breed):
        # super().__init__(name)  ← forgot this!
        self.breed = breed

    def describe(self):
        print(f"{self.name} is a {self.breed}")  # AttributeError: 'Dog' object has no attribute 'name'

dog = Dog("Buddy", "Labrador")
# dog.describe()  # Uncomment to see the error

# Good: always call super().__init__ first
class GoodDog(Animal):
    def __init__(self, name, breed):
        super().__init__(name)   # parent attribute created ✓
        self.breed = breed

    def describe(self):
        print(f"{self.name} is a {self.breed}")

gdog = GoodDog("Buddy", "Labrador")
gdog.describe()  # Works correctly
````

## Practical Example: Vehicle Hierarchy

````python
class Vehicle:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year
    
    def start(self):
        print(f"{self.brand} is starting...")

class Car(Vehicle):
    def __init__(self, brand, year, num_doors):
        super().__init__(brand, year)
        self.num_doors = num_doors
    
    def honk(self):
        print("Beep beep!")

class Motorcycle(Vehicle):
    def __init__(self, brand, year, has_sidecar):
        super().__init__(brand, year)
        self.has_sidecar = has_sidecar
    
    def wheelie(self):
        print("Doing a wheelie!")

car = Car("Toyota", 2020, 4)
bike = Motorcycle("Harley", 2019, False)

car.start()   # Inherited
car.honk()    # Car-specific

bike.start()  # Inherited
bike.wheelie()  # Motorcycle-specific
````

# Part 7: When to Use (and Avoid) OOP

## When to Use Classes

**✓ Use classes when**:
1. You have related data and behaviors
2. You need multiple instances of the same "thing"
3. You want to organize complex systems
4. Code reuse through inheritance makes sense

**Examples**:
- User management system (User class)
- Game objects (Player, Enemy, Item)
- Data models (Product, Order, Customer)
- UI components (Button, TextField, Window)

## When to Avoid Classes

Classes shine when you need **multiple objects that carry state and behaviour**. They add overhead in simpler situations.

**Avoid classes when**:
| Situation | Better alternative |
|-----------|-------------------|
| A short script with a handful of functions | Module-level functions |
| Data transformation pipeline (input → output) | Plain functions / generators |
| Behaviour never needs to vary across instances | Module-level constants + functions |
| Your class hierarchy is growing beyond 2–3 levels deep | Flatten the hierarchy; favour composition |

**Quick test**: "Would I ever create two separate instances with different state?"
- *Yes* → a class is probably a good fit.
- *No* → start with functions; add a class later if complexity grows.

## Gotcha #39: Over-Engineering — Not Every Problem Needs a Class

Don't reach for a class just because you can. Stateless utility operations are cleaner as plain functions:

````python
# Bad: Unnecessary class
class MathOperations:
    def add(self, a, b):
        return a + b
    
    def subtract(self, a, b):
        return a - b

# Good: Just use functions
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b
````

**Rule of thumb**: If you don't need multiple instances with state, don't use a class.

## Composition: "Has-A" Relationships

When two classes are related but one is not a *type* of the other, use **composition** — store one object as an attribute of the other. This is often more flexible than inheritance because you can swap out components independently without restructuring a class hierarchy.

````python
# Composition: A Car HAS-AN Engine
class Engine:
    def __init__(self, horsepower):
        self.horsepower = horsepower
    
    def start(self):
        print("Engine starting...")

class Car:
    def __init__(self, brand, horsepower):
        self.brand = brand
        self.engine = Engine(horsepower)  # Composition
    
    def start(self):
        self.engine.start()
        print(f"{self.brand} is ready to go!")

car = Car("Toyota", 200)
car.start()
````

**When in doubt, prefer composition over inheritance.** Inheritance encodes an IS-A relationship in the class hierarchy permanently — it's hard to undo later. Composition (HAS-A) is easier to change: swap the `Engine` object in a `Car` without touching any other class. Favour inheritance when the IS-A relationship is genuinely stable and meaningful.

| 38 | Forgetting `self` | Defining a method without `self` — Python tries to pass the instance but the method accepts no arguments, raising `TypeError` | Always include `self` as the first parameter in every instance method |

# Module 5 Complete Summary

## Classes and Objects
- **Class**: Blueprint for creating objects — define once, instantiate many times
- **Object (instance)**: A concrete object created from a class
- **`__init__`**: Constructor — runs automatically when you call `ClassName(...)`
- **`self`**: A reference to the specific instance the method is operating on

## Key Dunder Methods
- **`__str__`**: Called by `print()` and f-strings — human-readable representation
- **`__repr__`**: Called in the REPL and by `repr()` — developer/debugging representation; define this first

## Attributes
- **Instance attributes** (`self.x = ...` in `__init__`): each object gets its own copies
- **Class attributes** (defined directly in the class body): shared by all instances; modify via `ClassName.attr`, not `self.attr`
- **`@classmethod`**: method that receives the class (`cls`) rather than an instance; used for factory methods and class-level operations

## Inheritance and Polymorphism
- **Parent (base) class**: provides shared functionality
- **Child (derived) class**: inherits everything from the parent and can add or override
- **`super()`**: calls the parent class's version of a method; always use it in `__init__` to initialise parent attributes
- **Override**: redefine a parent method in the child
- **Polymorphism**: calling the same method name on different types automatically dispatches to the right implementation

## Design Principles
- Use classes for related data + behaviour that warrant multiple instances
- Prefer **composition** ("has-a") over deep inheritance chains when it keeps things simpler
- Keep each class focused on a single responsibility
- Don't over-engineer — plain functions are often the right tool
